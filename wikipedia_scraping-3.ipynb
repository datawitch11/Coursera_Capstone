#  Web scraping with beautiful soup and cleaning in form of desired table

## First, we import necessary libraries

```python

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import numpy as np 

%matplotlib inline
```

```python
from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup
```

## the URL for the page containing the table is read 

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

## pulling data out of HTML and XML files using beautiful soup library

```python
soup = BeautifulSoup(html, 'lxml')
type (soup)
```

```python
title=soup.title
print (title)
```

## Reading each row of the table 

```python
rows = soup.find_all('tr')
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

## extracting all the text in rows

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

## transforming list of rows into dataframe form

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

## now the list has to be separated by commas

```python
df1 = df[0].str.split(',',expand=True)
df1.head(10)
```

## Getting rid of unnecessary list characters such as [, ] and /n

```python
df1[0] = df1[0].str.strip('[')
df1[2] = df1[2].str.strip('\n]')
df1.head(10)
```

## filtering only first three columns

```python
df1.drop(df1.columns.difference([0,1,2]), 1, inplace=True)
```

```python
df1
```

## dropping rows that are not data from table

```python
df1.drop([0,289,290,291,292,293],inplace=True)
```

## deleting rows with 'Not assigned' borough

```python
df2=df1[~df1[1].str.contains("Not")]
```

# when we look at length of neighborhoods ( i.e df[3]) the length of the strings show that there is an extra space :

```python
df2[2].str.len()
```

## Getting rid of that extra space

```python
df2[2]=df2[2].str.strip()
```

## replacing the not assigned in neighborhood with its borough

```python
df2.replace({2: {"Not assigned": "Queen's Park"}})
```

## Now we group the data by postalcode and borough and combine all rows with same postalcode and borough

```python
df3 = df2.groupby([0, 1],as_index=False).agg(lambda x: x.tolist())
```

```python
df3
```

## I probably should have renamed columns earlier, but it was  a lot of work dealing with typing names. So, now the columns are renamed to proper names:

```python
df3.columns=['PostalCode','Borough','Neighborhood']
```

```python
df3
```

## our table is ready! here is the shape of it:

```python
df3.shape
```

## Let's read the CSV file containting geospatial coordinates 

```python
dfgeo=pd.read_csv('Geospatial_Coordinates.csv')
```

```python
dfgeo.head()
```

## The cell below loops through the scraped table and when it finds them in geospatial dataframe, it appends the corresponding latitude and longitude to latlistand longlist.

```python
latlist=[]
longlist=[]
for index, row in df3.iterrows():
        if (df3.loc[index, "PostalCode"]==dfgeo.loc[index,"Postal Code"]):
            latlist.append(dfgeo.loc[index,"Latitude"])
            longlist.append(dfgeo.loc[index,"Longitude"])
```

## To make sure the loop was correct, let's check if out latlist has the same number of elements as our scraped table (103)

```python
len(latlist)
```

## Perfect! Let's add latitude and longitude to our table 

```python
df3['Latitude']=latlist
```

```python
df3['Longitude']=longlist
```

```python
df3
```

## Now let's explore the neighborhoods containing the word "Toronto"


```python
df4= df3[df3['Borough'].str.contains("Toronto")]
```

```python
df4
```

```python
from geopy.geocoders import Nominatim
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

```python
import folium
# create map of New York using latitude and longitude values
map_Toronto= folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, postalcode in zip(df4['Latitude'], df4['Longitude'], df4['Borough'], df4['PostalCode']):
    label = '{}, {}'.format(borough, postalcode)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
